```python
import part_A as pa
import numpy as np
import pandas as pd

df=pd.read_csv("data.csv")
first_value = df.iloc[:100, 0].tolist()
second_value = df.iloc[:1000,1].tolist()
third_value =df.iloc[:10000,2].tolist()
print(first_value)
print(second_value)
print(third_value)
```

```python
#Timing calculation time using np
%timeit np.std(first_value)
%timeit np.std(second_value)
%timeit np.std(third_value)
```

```python
%timeit pa.std_loops(first_value)
%timeit pa.std_loops(second_value)
%timeit pa.std_loops(third_value)
```

```python
%timeit pa.std_builtin(first_value)
%timeit pa.std_builtin(second_value)
%timeit pa.std_builtin(third_value)
```

using np yields the fastest calculation time since it was built for vectorized operations, meaning it performs operations on the entire array at once instead of element after element like using for loops. That being said NumPy is the fastest in calculating the array with 10 000 numbers. The smaller the array gets NumPy becomes unsuitable, making the built in functions and for loops more suitable.
